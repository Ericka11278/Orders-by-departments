Goal analysis: Do people order different foods at different times of day?

order_hour_of_day: The hour of the day when the order was placed (0 = 12am, 1 = 1am etc.)
department: The department the products came from (e.g. alcohol, produce).
num_orders_hour: The number of products ordered from this particular department during this particular hour.
Tot_orders_dept: The total number of products ordered from this particular department across all hours of the day.

```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt

df = pd.read_csv('InstacartOrdersByDepartment.csv')
df.head()
```

build bar graphs for each department in which hour is on one axis and number of orders is on the other axis

```python
df
```

```python
import seaborn as sns
sns.relplot(
    data=df,
    x="order_hour_of_day", y="num_orders_hour", col="department")
```

In this seaborn plot, we can see the trends for each department. We can see all of the departments side by side and the amount of orders they had per hour which shows us which department is more popular and at which time. We can see that people purchase bakery, beverages, canned goods, eggs,frozen items, produce and snacks more during the day which hints to us that maybe that is the time people do more shopping and they perfer those items. We saw that those are the most popular ones during the day which means they are more prefered than other departments.
